```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- Caricamento file ---
file_path = '../Cone-Beam CT/Image_quality/Risultati_31_05_25/Testa_CAT_results.xlsx'
df = pd.read_excel(file_path)

# Crea la cartella "grafici" nella stessa directory del file, se non esiste
output_dir = os.path.join(os.path.dirname(file_path), 'grafici_Testa_CAT')
os.makedirs(output_dir, exist_ok=True)

# Crea le etichette x combinando Ref, kV, mA, mAs, CTDI (se non già presente)
if 'x_label' not in df.columns:
    df['x_label'] = df.apply(lambda row: f"{row['Ref']}\n{row['kV']}kV\n{row['mA']}mA\n{row['mAs']}mAs\nCTDI:{row['CTDI']}", axis=1)

def plot_group(columns, ylabel, title, filename, df=df):
    plt.figure(figsize=(12, 6))
    plotted = False
    for tip, group in df.groupby('Tipologia'):
        for col in columns:
            if col in group.columns and group[col].notna().any():
                plt.plot(group['x_label'], group[col], marker='o', label=f"{col} ({tip})")
                plotted = True
    plt.xlabel('Parametri')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.tight_layout()
    plt.grid(True)
    if plotted:
        plt.savefig(os.path.join(output_dir, filename), dpi=300)
    plt.close()

# 1. Sigma (ora 'Sigma')
plot_group(['Sigma'], 'Sigma', 'Andamento Sigma', 'sigma.png')

# 2. C_Tef e C_Poly
plot_group(['C_Tef', 'C_Poly'], 'Contrasto', 'Andamento C_Tef e C_Poly', 'C_Tef_C_Poly.png')

# 3. CNR_Tef, CNR_Poly
plot_group(['CNR_Tef', 'CNR_Poly'], 'CNR', 'Andamento CNR_Tef e CNR_Poly', 'CNR_Tef_CNR_Poly.png')

# 4. f_50_Tef, f_50_Poly, f10_Tef, f10_Poly
plot_group(['f_50_Tef', 'f_50_Poly', 'f10_Tef', 'f10_Poly'], 'Frequenza [mm$^{-1}$]', 'Andamento f50 e f10', 'f50_f10.png')

# 5. f_peak, f_av
plot_group(['f_peak', 'f_av'], 'Frequenza [mm$^{-1}$]', 'Andamento f_peak e f_av', 'f_peak_f_av.png')

# 6. d_NPW_Tef, d_NPW_Poly
plot_group(['d_NPW_Tef', 'd_NPW_Poly'], 'd_NPW', 'Andamento d_NPW_Tef e d_NPW_Poly', 'd_NPW_Tef_d_NPW_Poly.png')

# 7. Media_Tef, Mediana_Tef, Media_Poly, Mediana_Poly
plot_group(['Media_Tef', 'Mediana_Tef', 'Media_Poly', 'Mediana_Poly'], 'Valore', 'Media e Mediana', 'media_mediana.png')

# 8. Std_Tef, Std_Poly
plot_group(['Std_Tef', 'Std_Poly'], 'Deviazione standard', 'Std_Tef e Std_Poly', 'Std_Tef_Std_Poly.png')

# 9. Skwness_Tef, Kurtosi_Tef, Skwness_Poly, Kurtosi_Poly
plot_group(['Skwness_Tef', 'Kurtosi_Tef', 'Skwness_Poly', 'Kurtosi_Poly'], 'Valore', 'Skewness e Kurtosis', 'skewness_kurtosis.png')

print(f"Tutti i grafici sono stati salvati nella cartella: {output_dir}")
```
