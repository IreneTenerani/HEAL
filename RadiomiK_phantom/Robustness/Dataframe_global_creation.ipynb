```python
#Create daframe for the insert Cyl0
#Read in the csv files
#Load test-retest csv file into dataframe
df_cyl0_tr = pd.read_csv('../../../RadiomiK/Filtered_features/Test_ReTest/Final/Cyl0_testRetest.csv')

#Add a new column 'Test' to the dataframe in witch the value is 'Test-Retest' for all rows
df_cyl0_tr['Test'] = 'Test-Retest'

#Add a new column 'CT' to the dataframe in witch the value is 'TO-GE' for all rows
df_cyl0_tr['Manufacturer'] = 'TO-GE'

#delete the column 'count' from the dataframe
df_cyl0_tr = df_cyl0_tr.drop(['count'], axis=1)

#load the positioning csv file into a dataframe
df_cyl0_pos = pd.read_csv('../../../../RadiomiK/Features_Positioning_test/Features_Not_Repeatable/cv_Cyl0.csv')

#Add a new column 'Test' to the dataframe in witch the value is 'Positioning' for all rows
df_cyl0_pos['Test'] = 'Positioning'

#delete the column 'cv' from the dataframe
df_cyl0_pos = df_cyl0_pos.drop(['CV'], axis=1)

#Load the volume csv file into a dataframe
df_cyl0_vol = pd.read_csv('../../../../RadiomiK/Features_Volume_test/Features_Not_Repeatable/cv_Cyl0.csv')

#Add a new column 'Test' to the dataframe in witch the value is 'Volume' for all rows
df_cyl0_vol['Test'] = 'Volume'

#delete the column 'cv' from the dataframe
df_cyl0_vol = df_cyl0_vol.drop(['CV'], axis=1)

#Concatenate the three dataframes into one
df_cyl0 = pd.concat([df_cyl0_tr, df_cyl0_pos, df_cyl0_vol], ignore_index=True)

#Save the concatenated dataframe as csv file
df_cyl0.to_csv('../../../../RadiomiK/Features_not_repeatable_global/Cyl0_all_test.csv', index=False)

df_cyl0
```
