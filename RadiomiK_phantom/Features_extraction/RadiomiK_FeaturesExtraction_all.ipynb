```python
paramPath = 'Params_tol_0_Rad_001.yaml'
```

```python
############################################### da cambiare per ogni valore di dose e macchina
data_path = '../../../RadiomiK/Immagini/RADIOMIK_GE/RAD_GE/SHAPE_43/' #percorso cartella
mask_path = '../../../RadiomiK/Immagini/RADIOMIK_GE/RAD_GE_mask_water/SHAPE_43/' #percorso maschere di tutta la cartella

#data_path = '../../RadiomiK/Immagini/RADIOMIK_TO/RAD_TO/' #percorso cartella
#mask_path = '../../RadiomiK/Immagini/RADIOMIK_TO/RAD_TO_mask_water/' #percorso maschere di tutta la cartella

images_fnames = glob.glob(os.path.join(data_path,'*.dcm'))


############################################### da cambiare per ogni inserto
mask_name = 'water_43' #SPECIFICARE L'INSERTO: cyl0, cyl1, cyl2, cyl3, rect0, rect1, rect2, rect3, rect4, rect5, rect6, rect7, rect8, rect9, rect10, rect11, rect12, rect13, rect14, water, water_35, water_40, water_42, water_43
###############################################
masks_fnames = glob.glob(os.path.join(mask_path,'mask_' + mask_name + '*.dcm'))  

'''
mask_names = ['cyl0', 'cyl1', 'cyl2', 'cyl3', 'rect0', 'rect1', 'rect2', 'rect3', 'rect4', 'rect5', 'rect6', 'rect7', 'rect8', 'rect9', 'rect10', 'rect11', 'rect12', 'rect13', 'rect14']

masks_fnames_list = []

for mask_name in mask_names:
    masks_fnames = glob.glob(os.path.join(mask_path,'mask_' + mask_name + '*.dcm'))  
    masks_fnames_list.append(masks_fnames)
'''   
############################################### da cambiare per ogni dose e macchina
cwdfeatures = '../../../RadiomiK/Features_GE_water/' #cartella dove salvare le features 'Features_TO' , 'Features_TO_water', 'Features_GE', 'Features_GE_water'
###############################################

csv_file = os.path.join(cwdfeatures, 'Final/' + mask_name + '.csv')
```

```python
extractor = featureextractor.RadiomicsFeatureExtractor(paramPath)#, correctMask=True)    #verificare parametri
verbose = True
if verbose:
    print('Extraction parameters:\n\t', extractor.settings)
    print('Enabled filters:\n\t', extractor.enabledImagetypes)
    print('Enabled features:\n\t', extractor.enabledFeatures)
```

```python
#feature extraction
def_features = [{},{}]
restotal1=[0 for i in range(len(images_fnames))]
restotal=[]


for i in range(len(images_fnames)):
    
    imageName = images_fnames[i]
    maskNamepre = masks_fnames[0]
    filename = os.path.splitext(os.path.basename(imageName))[0]
    print(filename)
    
    num_acquisition = int(filename[-1])
    if num_acquisition > 3 or num_acquisition == 0:
        pass
    
    else:
   
        print("Calculating features")
        print(imageName)
        print(maskNamepre)
        image = pydicom.dcmread(imageName)
        seg = pydicom.dcmread(maskNamepre)
        # Convert the pydicom image to SimpleITK format
        image_sitk = sitk.GetImageFromArray(image.pixel_array)
        # Convert the pydicom segmentation mask to SimpleITK format
        seg_sitk = sitk.GetImageFromArray(seg.pixel_array)
        first_underscore_index = filename.find("_")
        second_underscore_index = filename.find("_", first_underscore_index + 1)
        third_underscore_index = filename.find("_", second_underscore_index + 1)
        fourth_underscore_index = filename.find("_", third_underscore_index + 1)
        fifth_underscore_index = filename.find("_", fourth_underscore_index + 1)
    
        result = extractor.execute(image_sitk, seg_sitk)
        result['image_ID'] = filename
        result['Manufacturer'] = image.Manufacturer
        result['Slice Thickness'] = image.SliceThickness
        result['Pixel_Spacing'] = image.PixelSpacing
        result['Kernel'] = filename[second_underscore_index +1 : third_underscore_index]
        result['Dose'] = filename[third_underscore_index + 1 : fourth_underscore_index]
        result['ASIR'] = filename[fourth_underscore_index +1 : fifth_underscore_index]
        result['Num_Acquisition'] = filename[-1]
        feat = []
        valore = []
        restotal.append(result)
        for key, value in result.items():
            feat.append(key)
            valore.append(value)
        def_features = {'features':feat, filename:valore}
        resframe=pd.DataFrame(def_features)
        restotal1[i]=pd.DataFrame(resframe)
        resframe.to_csv(os.path.join(cwdfeatures, mask_name + '_' + filename +'.csv'), index = False)  
        print('Computed ',i, '\n')
    
```
