```python
import pandas as pd
import pingouin as pg
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import statsmodels.api as sm
```

```python
# Loding the TO csv file
df = pd.read_csv("../../RadiomiK/Features_TO/Final/rect4.csv")
#Elimino le colonne diagnostics
string_to_match = 'diagnostics'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
string_to_match = 'Unnamed:'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
df_TO = df
df_TO
```

```python
# Loding the GE csv files and joining them in a single dataframe
df_21 = pd.read_csv("../../../RadiomiK/Features_GE/Final/rect4_21.csv")
df_35 = pd.read_csv("../../../RadiomiK/Features_GE/Final/rect4_35.csv")
df_40 = pd.read_csv("../../../RadiomiK/Features_GE/Final/rect4_40.csv")
df_42 = pd.read_csv("../../../RadiomiK/Features_GE/Final/rect4_42.csv")
df_43 = pd.read_csv("../../../RadiomiK/Features_GE/Final/rect4_43.csv")

df = pd.concat([df_21, df_35, df_40, df_42, df_43], ignore_index=True)

#Elimino le colonne diagnostics
string_to_match = 'diagnostics'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
string_to_match = 'Unnamed:'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
df_GE = df
df_GE
```

```python
df_TOGE = pd.concat([df_TO, df_GE], ignore_index=True)
df_TOGE
```

```python
#Save the dataframe in a csv file
df_TOGE.to_csv('../../../RadiomiK/Features_Global/rect4_glob.csv', index=False)
```
