```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Crea una tabella pivot con 'Feature' come indice, 'New_Insert' come colonne e conta le occorrenze
pivot_df = df_GE.pivot_table(index='Feature', columns='New_insert', aggfunc='size', fill_value=0)

# Ordina il DataFrame in base alla somma delle righe, dal più grande al più piccolo
pivot_df = pivot_df.reindex(pivot_df.sum(axis=1).sort_values(ascending=False).index)

#Reorder the columns of the dataframe
pivot_df = pivot_df[['cyl0', 'cyl1', 'cyl3', 'rect0_1R', 'rect3_1R', 'rect6_2R', 'rect12_2R', 'rect13_2R', 'rect14_3R']]

color_map_insert = {
    'cyl0': plt.cm.Reds(0.3),
    'cyl1': plt.cm.Reds(0.5),
    'cyl2': plt.cm.Reds(0.7),
    'cyl3': plt.cm.Reds(0.9),
    'rect0_1R': plt.cm.Blues(0.3),
    'rect1_1R': plt.cm.Blues(0.5),
    'rect2_1R': plt.cm.Blues(0.7),
    'rect3_1R': plt.cm.Blues(0.9),
    'rect4_1R': plt.cm.Blues(1.1),
    'rect5_2R': plt.cm.Greens(0.3),
    'rect6_2R': plt.cm.Greens(0.5),
    'rect7_2R': plt.cm.Greens(0.7),
    'rect8_2R': plt.cm.Greens(0.9),
    'rect9_2R': plt.cm.Greens(1.1),
    'rect10_2R': plt.cm.Greys(0.25),
    'rect11_2R': plt.cm.Greys(0.4),
    'rect12_2R': plt.cm.Greys(0.6),
    'rect13_2R': plt.cm.Greys(0.8),
    'rect14_3R': plt.cm.Greys(1.0)
}


# Crea una lista di colori per ogni insert
colors = plt.cm.Reds(np.linspace(0.3, 1, len(pivot_df.columns[pivot_df.columns.str.contains('cyl')])))
colors = np.concatenate([colors, plt.cm.Blues(np.linspace(0.3, 1, len(pivot_df.columns[pivot_df.columns.str.contains('1R')])))])
colors = np.concatenate([colors, plt.cm.Greens(np.linspace(0.3, 1, len(pivot_df.columns[pivot_df.columns.str.contains('2R')])))])
colors = np.concatenate([colors, plt.cm.Greys(np.linspace(0.4, 1, len(pivot_df.columns[pivot_df.columns.str.contains('3R')])))])
# Color map personalizzata con color_map_insert
colors = color_map_insert['cyl0'], color_map_insert['cyl1'], color_map_insert['cyl3'], color_map_insert['rect0_1R'], color_map_insert['rect3_1R'], color_map_insert['rect6_2R'], color_map_insert['rect12_2R'], color_map_insert['rect13_2R'], color_map_insert['rect14_3R']


# Crea un grafico a barre impilate con colori personalizzati
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)

plt.xlabel('Features')
plt.ylabel('Counts')
plt.title('Number of times each feature has CV > 10%, by insert for GE')
plt.xticks(rotation=90)  # Ruota le etichette x per una migliore leggibilità
plt.legend(loc='upper right')  # Cambia la posizione della legenda

plt.show()

```
