```python
import pandas as pd
import pingouin as pg
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

import statsmodels.api as sm
```

```python
# Loding the TO csv file
df1 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/TO_test/Final/Test_100/cyl0.csv")
df2 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/TO_test/Final/Test_75/cyl0.csv")
df3 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/TO_test/Final/Test_50/cyl0.csv")


# Concatenating the 10 csv files
df = pd.concat([df1, df2, df3])

#Elimino le colonne diagnostics
string_to_match = 'diagnostics'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
string_to_match = 'Unnamed:'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
df_TO = df
df_TO
```

```python
# Loding the GE csv files and joining them in a single dataframe
df1 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/GE_test/Final/Test_100/cyl0.csv")
df2 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/GE_test/Final/Test_75/cyl0.csv")
df3 = pd.read_csv("../../../../RadiomiK/Features_Volume_test/GE_test/Final/Test_50/cyl0.csv")


df = pd.concat([df1, df2, df3])

#Elimino le colonne diagnostics
string_to_match = 'diagnostics'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
string_to_match = 'Unnamed:'
filtered_columns = [col for col in df.columns if string_to_match not in col]
df = df[filtered_columns]
df_GE = df
df_GE
```

```python
#Save the dataframe in a csv file
df_TO.to_csv('../../../../RadiomiK/Features_Volume_test/TO_test/Final/Global/cyl0_glob.csv', index=False)
df_GE.to_csv('../../../../RadiomiK/Features_Volume_test/GE_test/Final/Global/cyl0_glob.csv', index=False)
```

```python
#add a column to the dataframe with the number of test that goes from 1 to 3
df_TO.insert(0, 'Test', range(1, 1 + len(df_TO)))
df_GE.insert(0, 'Test', range(1, 1 + len(df_GE)))

df_TO
```

```python
#Keep only the [1,87] columns of the dataframe
df_TO = df_TO.iloc[:, 1:87]
df_GE = df_GE.iloc[:, 1:87]

df_TO
```

```python
#calculate the CV of each feature
cv_TO = df_TO.std() / df_TO.mean()
cv_GE = df_GE.std() / df_GE.mean()

#Create a dataframe with the CV of each feature with the name of the column "CV"
cv_TO = pd.DataFrame(cv_TO, columns=['CV'])
cv_GE = pd.DataFrame(cv_GE, columns=['CV'])

#add column with the name of the feature
cv_TO['Feature'] = cv_TO.index
cv_GE['Feature'] = cv_GE.index

#change the index of the dataframe
cv_TO = cv_TO.reset_index(drop=True)
cv_GE = cv_GE.reset_index(drop=True)

#Add a column Manufacturer with the name of the manufacturer
cv_TO['Manufacturer'] = 'Toshiba'
cv_GE['Manufacturer'] = 'GE'

#concatenate the two dataframe
cv = pd.concat([cv_TO, cv_GE])

#reset the index
cv = cv.reset_index(drop=True)


cv
```

```python
#Get the features with CV > 0.10
cv_10 = cv[cv['CV'] > 0.10]

#reset the index
cv_10 = cv_10.reset_index(drop=True)

cv_10
```

```python
#Save the dataframe in a csv file
cv_10.to_csv('../../../../RadiomiK/Features_Volume_test/Features_Not_repeatable/cv_Cyl0.csv', index=False)
```
