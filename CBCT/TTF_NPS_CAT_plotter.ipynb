```python
import pandas as pd
import matplotlib.pyplot as plt
import os

base_dir = '../Cone-Beam CT/Image_quality/Risultati_31_05_25/Testa_High_Catphan/Seconda_acquisizione/standard_2.5'

# Percorsi dei file
file_poly = os.path.join(base_dir, 'TTF_Poly.csv')
file_teflon = os.path.join(base_dir, 'TTF_Tef.csv')
file_nps = os.path.join(base_dir, 'NPS.csv')

# Carica i dati
df_nps = pd.read_csv(file_nps)
# Carica i dati
df_poly = pd.read_csv(file_poly)
df_teflon = pd.read_csv(file_teflon)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df_poly['f'], df_poly['TTF'], marker='o', color = 'b', label='Polystyrene')
plt.plot(df_teflon['f'], df_teflon['TTF'], marker='s', color='g', label='Teflon')
plt.xlabel('Frequenza [mm$^{-1}$]')
plt.ylabel('TTF')
plt.title('TTF for Polystyrene e Teflon')
plt.legend()
plt.grid(True)
plt.tight_layout()

output_dir = os.path.dirname(file_poly)
output_path = os.path.join(output_dir, 'TTF_Poly_Teflon.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot salvato in: {output_path}")

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df_nps['f'], df_nps['NPS'], marker='o', label='NPS')
plt.xlabel('Frequenza [mm$^{-1}$]')
plt.ylabel('NPS [HU$^2$ mm$^2$]')
plt.title('NPS')
plt.legend()
plt.grid(True)
plt.tight_layout()

output_dir = os.path.dirname(file_nps)
output_path = os.path.join(output_dir, 'NPS.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot salvato in: {output_path}")

```
