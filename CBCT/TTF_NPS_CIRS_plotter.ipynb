```python
import pandas as pd
import matplotlib.pyplot as plt
import os

base_dir = '../Cone-Beam CT/Image_quality/Risultati_31_05_25/Addome1/6-s/iCBCT4'

# Percorsi dei file
file_1L = os.path.join(base_dir, 'TTF_1L.csv')
file_2M = os.path.join(base_dir, 'TTF_2M.csv')
file_3H = os.path.join(base_dir, 'TTF_3H.csv')
file_nps = os.path.join(base_dir, 'NPS.csv')

# Carica i dati
df_nps = pd.read_csv(file_nps)
# Carica i dati
df_1L = pd.read_csv(file_1L)
df_2M = pd.read_csv(file_2M)
df_3H = pd.read_csv(file_3H)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df_1L['f'], df_1L['TTF'], marker='o', label='TTF 1L')
plt.plot(df_2M['f'], df_2M['TTF'], marker='o', label='TTF 2M')
plt.plot(df_3H['f'], df_3H['TTF'], marker='o', label='TTF 3H')
plt.xlabel('Frequenza [mm$^{-1}$]')
plt.ylabel('TTF')
plt.title('TTF')
plt.legend()
plt.grid(True)
plt.tight_layout()

output_dir = os.path.dirname(file_1L)
output_path = os.path.join(output_dir, 'TTF_Poly_Teflon.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot salvato in: {output_path}")

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df_nps['f'], df_nps['NPS'], marker='o', label='NPS')
plt.xlabel('Frequenza [mm$^{-1}$]')
plt.ylabel('NPS [HU$^2$ mm$^2$]')
plt.title('NPS')
plt.legend()
plt.grid(True)
plt.tight_layout()

output_dir = os.path.dirname(file_nps)
output_path = os.path.join(output_dir, 'NPS.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot salvato in: {output_path}")

```
